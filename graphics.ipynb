```python
#1a
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv')
```

```python
#1b
import datetime

time_year = [0] * df.shape[0]
for i in range(df.shape[0]):
    time_year[i] = datetime.datetime.strptime(df['Date'][i], "%Y-%m-%d")

df['time_year'] = time_year
```

```python
#1c Line plot
import matplotlib.pyplot as plt

fig, ax = plt.subplots(num=None, figsize=(10,5), dpi=100, facecolor="w")
plt.plot(df['time_year'], df['Temp'])
plt.title('Daily minimum temperature in Melbourne')
plt.grid(True)
plt.xlabel('Year')
plt.ylabel('Temperature')
ax.set_xlim([datetime.date(1981,1,1),datetime.date(1991,1,1)])
ax.set_ylim([0,25])
plt.show()
```

```python
#1c Scatter plot
fig, ax = plt.subplots(num=None, figsize=(10,5), dpi=100, facecolor="w")
plt.scatter(df['time_year'], df['Temp'])
plt.title('Daily minimum temperature in Melbourne')
plt.grid(True)
plt.xlabel('Year')
plt.ylabel('Temperature')
ax.set_xlim([datetime.date(1981,1,1),datetime.date(1991,1,1)])
ax.set_ylim([0,25])
plt.show()
```

```python
#1c Bar plot

fig, ax = plt.subplots(num=None, figsize=(10,5), dpi=100, facecolor="w")
plt.bar(df['time_year'], df['Temp']) 
plt.title('Daily minimum temperature in Melbourne')
plt.grid(True)
plt.xlabel('Year')
plt.ylabel('Temperature')
ax.set_xlim([datetime.date(1981,1,1),datetime.date(1991,1,1)])
ax.set_ylim([0,25])
plt.show()
```

```python
#1d Histogram
import numpy as np
plt.figure(num=None, figsize=(12,4), dpi=100, facecolor="w")
plt.hist(df['Temp'], bins = np.arange(0,27,1))
plt.xlabel('Temperature Celcius')
plt.grid(True)
plt.ylabel('Number of days')
plt.title('Daily minimum temperature in Melbourne')
plt.show()
```

```python
#2
data = pd.read_csv('mtcars.csv')
fig, ax = plt.subplots(num=None, figsize=(10,10), dpi=100, facecolor="w")
plt.subplot(2,2,1)
plt.hist(data['mpg'], bins=np.arange(10, 35, 1))
plt.xlabel('Miles per gallon')
plt.ylabel('Count')

plt.subplot(2,2,2)
plt.hist(data['cyl'], bins=np.arange(4, 10, 1))
plt.xlabel('Number of cylinders')
plt.ylabel('Count')

plt.subplot(2,2,3)
plt.hist(data['hp'])
plt.xlabel('Horsepower')
plt.ylabel('Count')

plt.subplot(2,2,4)
plt.hist(data['gear'], bins=np.arange(3, 7, 1))
plt.xlabel('Gear')
plt.ylabel('Count')
plt.show()
```
